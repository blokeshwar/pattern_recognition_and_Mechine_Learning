```python
import numpy as np

def get_input_matrix(rows, columns, matrix_num):
    matrix = []
    for i in range(rows):
        row = list(map(int, input(f"Enter row {i + 1} of matrix {matrix_num} (separate elements by space): ").split()))
        if len(row) != columns:
            print("Invalid number of elements in the row. Try again.")
            return None
        matrix.append(row)
    return np.array(matrix)

def matrix_addition(matrix1, matrix2):
    return np.add(matrix1, matrix2)

def matrix_subtraction(matrix1, matrix2):
    return np.subtract(matrix1, matrix2)

def scalar_matrix_multiplication(matrix, scalar):
    return np.multiply(matrix, scalar)

def elementwise_matrix_multiplication(matrix1, matrix2):
    return np.multiply(matrix1, matrix2)

def matrix_multiplication(matrix1, matrix2):
    return np.dot(matrix1, matrix2)

def matrix_transpose(matrix):
    return np.transpose(matrix)

def trace_of_matrix(matrix):
    return np.trace(matrix)

def solve_linear_equations(matrix, vector):
    return np.linalg.solve(matrix, vector)

def matrix_determinant(matrix):
    return np.linalg.det(matrix)

def matrix_inverse(matrix):
    return np.linalg.inv(matrix)

def eigen_values_and_vectors(matrix):
    return np.linalg.eig(matrix)

def main():
    # Getting inputs from the user
    m1 = int(input("Enter the number of rows for Matrix 1: "))
    n1 = int(input("Enter the number of columns for Matrix 1: "))
    matrix1 = get_input_matrix(m1, n1, 1)

    m2 = int(input("Enter the number of rows for Matrix 2: "))
    n2 = int(input("Enter the number of columns for Matrix 2: "))
    matrix2 = get_input_matrix(m2, n2, 2)

    if matrix1 is None or matrix2 is None:
        print("Invalid input. Exiting...")
        return

    print("\nInputs:")
    print("Matrix 1:")
    print(matrix1)
    print("Matrix 2:")
    print(matrix2)

    # Menu driven function
    while True:
        print("\nMatrix Operations:")
        print("1. Matrix Addition")
        print("2. Matrix Subtraction")
        print("3. Scalar Matrix Multiplication")
        print("4. Elementwise Matrix Multiplication")
        print("5. Matrix Multiplication")
        print("6. Matrix Transpose")
        print("7. Trace of a Matrix")
        print("8. Solve System of Linear Equations")
        print("9. Determinant")
        print("10. Inverse")
        print("11. Eigen Value and Eigen Vector")
        print("12. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            print("Matrix Addition:")
            result = matrix_addition(matrix1, matrix2)
            print(result)

        elif choice == 2:
            print("Matrix Subtraction:")
            result = matrix_subtraction(matrix1, matrix2)
            print(result)

        elif choice == 3:
            scalar = float(input("Enter the scalar value: "))
            print("Scalar Matrix Multiplication:")
            rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2\n"))
            if rot == 1:
                result = scalar_matrix_multiplication(matrix1, scalar)
            elif rot == 2:
                result = scalar_matrix_multiplication(matrix2, scalar)
            print(result)

        elif choice == 4:
            print("Elementwise Matrix Multiplication:")
            result = elementwise_matrix_multiplication(matrix1, matrix2)
            print(result)

        elif choice == 5:
            print("Matrix Multiplication:")
            result = matrix_multiplication(matrix1, matrix2)
            print(result)

        elif choice == 6:
            print("Matrix Transpose:")
            rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2 \n"))
            if rot == 1:
                result = matrix_transpose(matrix1)
            elif rot == 2:
                result = matrix_transpose(matrix2)
            print(result)

        elif choice == 7:
            print("Trace of a Matrix:")
            rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2 \n"))
            if rot == 1:
                result = trace_of_matrix(matrix1)
            elif rot == 2:
                result = trace_of_matrix(matrix2)
            print(result)

        elif choice == 8:
            vector_size = m1  # Use the number of rows in Matrix 1 as vector size
            vector = np.array(list(map(float, input(f"Enter the vector of size {vector_size} (separate elements by space): ").split())))
            if len(vector) != vector_size:
                print("Invalid vector length!")
            else:
                rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2 \n"))
                if rot == 1:
                    result = solve_linear_equations(matrix1, vector)
                elif rot == 2:
                    result = solve_linear_equations(matrix2, vector)
                print(result)

        elif choice == 9:
            print("Determinant:")
            rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2 \n"))
            if rot == 1:
                result = matrix_determinant(matrix1)
            elif rot == 2:
                result = matrix_determinant(matrix2)
            print(result)

        elif choice == 10:
            print("Inverse:")
            rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2 \n"))
            if rot == 1:
                result1 = matrix_determinant(matrix1)
                if result1 != 0:
                    result = matrix_inverse(matrix1)
                else:
                    print("No inverse for this matrix")
            elif rot == 2:
                result1 = matrix_determinant(matrix2)
                if result1 != 0:
                    result = matrix_inverse(matrix2)
                else:
                    print("No inverse for this matrix")
            print(result)

        elif choice == 11:
            print("Eigen Value and Eigen Vector:")
            rot = int(input("Choose the matrix \n 1. Matrix 1 \n 2. Matrix 2 \n"))
            if rot == 1:
                eigen_values, eigen_vectors = eigen_values_and_vectors(matrix1)
            elif rot == 2:
                eigen_values, eigen_vectors = eigen_values_and_vectors(matrix2)
            print("Eigen Values:")
            print(eigen_values)
            print("\nEigen Vectors:")
            print(eigen_vectors)

        elif choice == 12:
            print("Exiting...\n")
            break

        else:
            print("Invalid choice! Try again.")

if __name__ == "__main__":
    main()
12
```
