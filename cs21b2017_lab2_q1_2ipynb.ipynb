```python
df_test=pd.DataFrame(columns=df.columns)
df_train=pd.DataFrame(columns=df.columns)
for i in df['Species'].unique():
  df1=df.groupby('Species')
  df1=df1.get_group(i)
  df1=df1.iloc[0:10]
  df2=df.groupby('Species')
  df2=df2.get_group(i)
  df2=df2.iloc[10:len(df2)]
  df_train=pd.concat([df_train,df2])
  df_test=pd.concat([df_test,df1])
```

```python
X_train=df_train.drop(['Species'],axis=1)
y_train=df_train['Species']
X_test=df_test.drop(['Species'],axis=1)
y_test=df_test['Species']
```

```python
X_train=X_train.to_numpy()
```

```python
X_test=X_test.to_numpy()
```

```python
from math import sqrt
from collections import Counter
def Euclidean_distance(X, Y):
    sum_sq = np.sum(np.square(X-Y))
    return sqrt(sum_sq)
```

```python
def knn_predict(X_train, X_test, y_train, y_test, k):
    y_hat = []

    for test_point in X_test:
        distances = []

        for train_point in X_train:
            distance = Euclidean_distance(test_point, train_point)
            distances.append(distance)
        df_dists = pd.DataFrame(data=distances, columns=['dist'],
                                index=y_train.index)
        df_nn = df_dists.sort_values(by=['dist'], axis=0)[:k]
        counter = Counter(y_train[df_nn.index])
        prediction = counter.most_common()[0][0]
        y_hat.append(prediction)

    return y_hat
```

```python
y1=knn_predict(X_train,X_test,y_train,y_test,k=1)
y3=knn_predict(X_train,X_test,y_train,y_test,k=3)
y5=knn_predict(X_train,X_test,y_train,y_test,k=5)
y6=knn_predict(X_train,X_test,y_train,y_test,k=6)
y7=knn_predict(X_train,X_test,y_train,y_test,k=7)
print("for k=1:",y1)
print("for k=3:",y3)
print("for k=5:",y5)
print("for k=6:",y6)
print("for k=7:",y7)
```

```python
y1 = list(map(int,y1))
```

```python
y_test = list(map(int,y_test))
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy for k=1:",accuracy_score(y1,y_test))
print("The accuracy for k=3:",accuracy_score(y3,y_test))
print("The accuracy for k=5:",accuracy_score(y5,y_test))
print("The accuracy for k=6:",accuracy_score(y6,y_test))
print("The accuracy for k=7:",accuracy_score(y7,y_test))
```
